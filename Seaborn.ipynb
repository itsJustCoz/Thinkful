```python
#lmplot
g = sns.lmplot(y='fixed acidity', 
               x='quality',
               data=redwine, 
               fit_reg=True, 
               scatter_kws={'alpha':0.4})
g.set_ylabels("Fixed Acidity")
g.set_xlabels("Quality")
plt.title('Scatterplot with regression line: Fixed Acidity by Quality')
plt.show()
print("Easy to see relationship between data and frequency of data. Hard to see the shape of the dataset")

#jointplot
sns.jointplot(y='fixed acidity', x='quality', data=redwine, kind="kde")
plt.title('Joint Plot')
plt.show()
print("Easy to see the shape and measures of central tendency and relationship of variables, but not as easy to see the relationship as the scatterplot")

#KDE plot different density
f, ax = plt.subplots(figsize=(6, 6))
cmap = sns.cubehelix_palette(as_cmap=True, dark=0, light=1, reverse=True)
sns.kdeplot(redwine.quality, redwine['fixed acidity'], cmap=cmap, n_levels=60, shade=True);
plt.title('KDE plt')
plt.show()
print("Bad at almost everything. It's easiest to see where the quality and fixed acidity group")
```

```python
divvy = pd.read_csv("Chicago-Divvy-2016.csv")
# convert seconds to minutes
divvy['tripduration'] = divvy['tripduration']//60
# add a day column
divvy['weekday'] = divvy['starttime'].apply(lambda x: pd.to_datetime(x)).dt.weekday_name
# Sort weekday column
divvy['weekday'] = pd.Categorical(divvy['weekday'], categories=['Sunday', 'Saturday', 'Friday', 'Thursday', 'Wednesday', 'Tuesday', 'Monday'], ordered=True)
```

```python
#stripplot
f, ax = plt.subplots(figsize=(8, 6))
sns.stripplot(x="weekday", y="tripduration", data=divvy)
plt.show()
print('Handles outliers well, but the frequency of the datapoints is impossible to see. No measures of central tendency.')

#striplot jittered
f, ax = plt.subplots(figsize=(8, 6))
sns.stripplot(x="weekday", y="tripduration", data=divvy, jitter=True)
plt.show()
print('Handles outliers well, and you regain the sense of frequency in the lowere range of the data. No measures of central tendency.')

#Boxplot limited axis
f, ax = plt.subplots(figsize=(8, 6))
sns.boxplot(x="weekday", y="tripduration", data=divvy, dodge=False)
plt.ylim(0, 100)
plt.show()
print('Handles extreme outliers very poorly. Had to limit the Y axis. Show the measueres of central tendency very well. Doesn\'t show the shape of the data well.')

#Violin plot
f, ax = plt.subplots(figsize=(8, 6))
sns.violinplot(x="weekday", y="tripduration", data=divvy)
plt.ylim(0, 100)
plt.show()
print("Same strengths and weakenesses as the boxplot,but it shows the \"with\" of the data better than the boxplot.")

#barplot
f, ax = plt.subplots(figsize=(8, 6))
sns.barplot(x="weekday", y="tripduration", data=divvy)
plt.show()
print("Easy to see how the mean varies by day, but you get no sense of the shape of the data.")      

#pointplot
f, ax = plt.subplots(figsize=(8, 6))
sns.pointplot(x="weekday", y="tripduration", data=divvy)
plt.show()
print("Same as above, but at least you get a slightly better idea of variance.")
```
